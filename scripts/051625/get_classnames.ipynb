```python
import os
import json
import numpy as np

# os.chdir("/home/zhang.14217/bioclip-dev")
os.chdir("#### YOUR WORKING DIRECTORY ####")

train_path = "config/data/MAD_MAD05_new/train.json"
test_path = "config/data/MAD_MAD05_new/test.json"

train_json = json.load(open(train_path))
test_json = json.load(open(test_path))
```

```python
all_class_names = set()

for ckp in train_json:
    ckp_d = train_json[ckp]
    for d in ckp_d:
        class_name = d["class_name"]
        all_class_names.add(class_name)

for ckp in test_json:
    ckp_d = test_json[ckp]
    for d in ckp_d:
        class_name = d["class_name"]
        all_class_names.add(class_name)
```

```python
for n in all_class_names:
    print(f"  - {n}")
```
